```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier
from difflib import SequenceMatcher
from sklearn import svm
```

```python
def returner(df):
    return df
```

```python
spotData = pd.read_json("spotData5.json")
```

```python
a = {}
counter = 1
for i in set(spotData["playlist"]):
    a[i] = counter
    counter+=1
```

```python
floating = []
for row in spotData["playlist"]:
    floating.append(a[row])
```

```python
spotData["playlist"] = floating
```

```python
returner(spotData)
```

```python
x_train, x_test, y_train, y_test = train_test_split(spotData[["playlist","danceability","energy","key","loudness","mode","speechiness","acousticness","instrumentalness","valence","tempo"]], spotData["mood"], test_size=.25, random_state=0)

extraTreesModel = ExtraTreesClassifier(n_estimators = 100)
extraTreesClassified = extraTreesModel.fit(x_train, y_train)
predictions = extraTreesModel.predict(x_test)

Accuracy = extraTreesModel.score(x_test, y_test)
print("Test Accuracy: {:.4f}".format(Accuracy*100))
```
